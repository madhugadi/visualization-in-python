```python
traffic=pd.read_csv('traffic_volume.csv')
```

```python
traffic.head()
```

```python
traffic.keys()
```

```python
traffic['date_time']=pd.to_datetime(traffic['date_time'])
```

```python
traffic['year']=traffic['date_time'].dt.year
```

```python
traffic['month']=traffic['date_time'].dt.month
```

```python
traffic['day']=traffic['date_time'].dt.day
```

```python
month_wise=traffic.groupby(['year','month'])['traffic_volume'].mean().reset_index()
```

```python
#traffic.to_csv('traffic_month_wise.csv')
```

```python
traffic.head()
```

```python
month_wise
```

```python
month_wise1=month_wise.pivot('month','year','traffic_volume')
month_wise1
```

```python
sns.heatmap(month_wise1.corr(),annot=True)
```

```python
day_wise=traffic.groupby(['month','day'])['traffic_volume'].mean().reset_index()
```

```python
day_wise1=day_wise.pivot('month','day','traffic_volume')
```

```python
sns.heatmap(day_wise1)
```

```python
import sounddevice as sd
import numpy as np
import scipy.io.wavfile as wav     ## channel1 - monorecording,channel2-stereo recording
fs=8000
duration=10 # seconds
myrecording=sd.rec(duration*fs,samplerate=fs,channels=1,dtype='float64')
print("Recording audio")
sd.wait()
print("Audio recording complete, Play Audio")
```

```python
sd.play(myrecording,10000)
sd.wait()
print("Play Audio complete")
```
